```python
import pandas as pd 

df_train=pd.read_csv("/content/drive/MyDrive/train.csv")
df_test=pd.read_csv("/content/drive/MyDrive/test.csv")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
import numpy as np
```

```python
df_train.head(15)
```

```python
list_of_all_ages = df_train["Age"].tolist()
```

```python
print(list_of_all_ages)
```

```python
list_of_ages_without_nan = [x for x in list_of_all_ages if str(x) != "nan"]
```

```python
print(list_of_ages_without_nan)
```

```python
def mean_age(list_of_ages):
  sum_of_ages = sum(list_of_ages)
  mean_of_ages = sum_of_ages / len(list_of_ages)
  return round(mean_of_ages, 0)
```

```python
mean = mean_age(list_of_ages_without_nan)
print(mean)
```

```python
df_train["Age"] = df_train["Age"].fillna(mean)
```

```python
df_train.head(6)
```

```python
df_train["Sex"] = df_train["Sex"].replace("male", 0)
```

```python
df_train.head(5)
```

```python
df_train["Sex"] = df_train["Sex"].replace("female", 1)
```

```python
df_train.head(10)
```

```python
# Now the test data...

new_list_of_all_ages = df_test["Age"].tolist()
new_list_of_ages_without_nan = [x for x in list_of_all_ages if str(x) != "nan"]
new_mean = mean_age(new_list_of_ages_without_nan)
df_test["Age"] = df_test["Age"].fillna(new_mean)
df_test["Sex"] = df_test["Sex"].replace("male", 0)
df_test["Sex"] = df_test["Sex"].replace("female", 1)
df_test.head(20)




```

```python
train_data_x = df_train[["Pclass", "Sex", "Age"]]
train_data_x.head(10)
```

```python
train_data_y = df_train[["Survived"]]
train_data_y.head(10)
```

```python
test_data_x = df_test[["Pclass", "Sex", "Age"]]
test_data_x.head(30)
```

```python
model = LogisticRegression()
model.fit(train_data_x, np.ravel(train_data_y))
```

```python
y_predict = model.predict(test_data_x.iloc[[26]])
print(y_predict)
```

```python
predictions = []
```

```python
for idx in range(len(test_data_x)):
  y_predict = model.predict(test_data_x.iloc[[idx]])
  y_predict = int(y_predict)
  predictions.append(y_predict)
print(predictions)
print(len(predictions))
```

```python
print(type(predictions[0]))
```

```python
passengerids = df_test["PassengerId"].tolist()
print(passengerids)
print(len(passengerids))
```

```python
dictionary_of_lists = {"PassengerId": passengerids, "Survived": predictions}
final_dataframe = pd.DataFrame(dictionary_of_lists)
final_dataframe.head()
```

```python
final_dataframe.to_csv("predictions.csv", index=False)
```
